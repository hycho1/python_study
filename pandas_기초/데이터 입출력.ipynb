```python
# 위키피디아 미국 ETF 웹 페이지에서 필요한 정보를 스크래핑하여 딕셔너리 형태로 변수 etfs에 저장
resp = requests.get(url2)
soup = BeautifulSoup(resp.text, 'lxml')   
rows = soup.select('div > ul > li')
```

```python
rows
```

```python
etfs = {}
for row in rows:
    
    try:
        etf_name = re.findall('^(.*) \(NYSE', row.text)            # Apple (NYSE: AAPL) 에서 Apple과 NYSE
        etf_market = re.findall('\((.*)\|', row.text)              # (Hello|World) 에서 Hello 추출
        etf_ticker = re.findall('NYSE Arca\|(.*)\)', row.text)     # (text NYSE Arca|Example) 에서 Example 추출
        
        if (len(etf_ticker) > 0) & (len(etf_market) > 0) & (len(etf_name) > 0):
            etfs[etf_ticker[0]] = [etf_market[0], etf_name[0]]

    except AttributeError as err:
        pass    
```

```python
# 위키피디아 미국 ETF 웹 페이지에서 필요한 정보를 스크래핑하여 딕셔너리 형태로 변수 etfs에 저장
url = "https://en.wikipedia.org/w/index.php?title=List_of_American_exchange-traded_funds&oldid=1016743136"
resp = requests.get(url)
soup = BeautifulSoup(resp.text, 'lxml')   
rows = soup.select('div > ul > li')
```

```python
etfs = []
for row in rows:
    
    try:
        etf_name = re.findall('([^\n()]+)\s*\(', row.text)
        #etf_market = re.findall('NYSE', row.text)
        #etf_ticker = re.findall('\b([A-Za-z]{4})\b', row.text)
        
        if (len(etf_name) > 0):
            etfs.append(etf_name)

    except AttributeError as err:
        pass    
```

```python
# 판다스 DataFrame() 함수로 데이터프레임 변환. 변수 df1, df2에 저장 
data1 = {'name' : [ 'Jerry', 'Riah', 'Paul'],
         'algol' : [ "A", "A+", "B"],
         'basic' : [ "C", "B", "B+"],
          'c++' : [ "B+", "C", "C+"]}

data2 = {'c0':[1,2,3], 
         'c1':[4,5,6], 
         'c2':[7,8,9], 
         'c3':[10,11,12], 
         'c4':[13,14,15]}

df1 = pd.DataFrame(data1)
df1.set_index('name', inplace=True)      #name 열을 인덱스로 지정
df1
```

```python
df2 = pd.DataFrame(data2)
df2.set_index('c0', inplace=True)        #c0 열을 인덱스로 지정
df2
```

```python
# df1을 'sheet1'으로, df2를 'sheet2'로 저장 (엑셀파일명은 "df_excelwriter.xlsx")
writer = pd.ExcelWriter("./df_excelwriter.xlsx")
df1.to_excel(writer, sheet_name="sheet1")
df2.to_excel(writer, sheet_name="sheet2")
writer.save()
```
