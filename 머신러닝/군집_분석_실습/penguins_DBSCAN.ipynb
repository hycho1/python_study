```python
import pandas as pd
import numpy as np
import seaborn as sns

df = sns.load_dataset('penguins')   # 펭귄 dataset load

df.dropna(inplace=True) #df의 결측치들은 행째로 제거 (inplace=True를 통해 즉시 데이터에 적용)

# 수치 데이터만 저장한 데이터셋 생성
dataset = df[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(dataset)
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt


dbscan = DBSCAN(eps=0.55, min_samples=2) # DBSCAN (eps : epsilon, min_samples : min point)
dbscan.fit(scaled_data)
```

```python
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=dbscan.labels_) # Clustering 결과 시각화 용도
plt.show()
```
