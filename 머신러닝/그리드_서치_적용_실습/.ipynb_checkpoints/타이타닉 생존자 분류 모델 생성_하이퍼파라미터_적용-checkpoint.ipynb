- 로지스틱 회귀와 결정 트리 비교

## 데이터 준비
- 타이타닉

```python
import pandas as pd

titanic = pd.read_csv('titanic_train.csv')

titanic.head()
```

```python
titanic.info()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
```

## 불필요한 칼럼 제거

```python
titanic.drop(columns=['Cabin','Embarked','Ticket','Name'], inplace=True)
```

## 결측치 처리

```python
age_imputer = SimpleImputer(strategy='mean')
titanic['Age'] = age_imputer.fit_transform(titanic[['Age']])
titanic = pd.get_dummies(titanic, columns=['Sex'], drop_first=True)
```

```python
titanic
```

```python
titanic.describe()
```

```python
titanic.info()
```

```python
X = titanic[['Pclass','Age','Fare','SibSp','Parch','Sex_male']]
y = titanic['Survived'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

# train_input, test_input, train_target, test_target
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42) # test_size > 8:2로 트레인과 테스트를 분할
```

```python
X_train.shape, X_test.shape
```

#### 스케일링

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(X_train)

X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)
```

## 로지스틱 회귀 모델

#### 스케일링 O

```python
from sklearn.linear_model import LogisticRegression

lr_scaled = LogisticRegression()
lr_scaled.fit(X_train_scaled, y_train)

print(lr_scaled.score(X_train_scaled, y_train))
print(lr_scaled.score(X_test_scaled, y_test))
```

### 하이퍼 파라미터 적용

```python
## C (Regularization Strength) : C가 작을수록 규제가 강함

for i in [0.01, 0.1, 1, 10, 100]:
    lr_scaled_C = LogisticRegression(C=i)
    lr_scaled_C.fit(X_train_scaled, y_train)

    print("X_train C: ", i, "  \t SCORE :::: " , lr_scaled_C.score(X_train_scaled, y_train))
    print("X_test  C: ", i, "  \t SCORE :::: " , lr_scaled_C.score(X_test_scaled, y_test))
    print()
```

```python
##  penalty (Regularization Type)
##  l1 : L1 규제 (라쏘, Lasso)
##  l2 : L2 규제 (릿지, Ridge)
##  elasticnet : L1과 L2를 결합한 규제
##  none : 규제 없음

# L1 규제 모델
lr_scaled_penalty_l1 = LogisticRegression(penalty='l1', solver='liblinear')
lr_scaled_penalty_l1.fit(X_train_scaled, y_train)

print("X_train SCORE :::: " , lr_scaled_penalty_l1.score(X_train_scaled, y_train))
print("X_test  SCORE :::: " , lr_scaled_penalty_l1.score(X_test_scaled, y_test))
```

```python
# L2 규제 모델
lr_scaled_penalty_l2 = LogisticRegression(penalty='l2')
lr_scaled_penalty_l2.fit(X_train_scaled, y_train)

print("X_train SCORE :::: " , lr_scaled_penalty_l2.score(X_train_scaled, y_train))
print("X_test  SCORE :::: " , lr_scaled_penalty_l2.score(X_test_scaled, y_test))
```

```python
# Elastic Net 규제 모델
lr_scaled_penalty_elasticnet = LogisticRegression(penalty='elasticnet', solver='saga', l1_ratio=0.5)
lr_scaled_penalty_elasticnet.fit(X_train_scaled, y_train)

print("X_train SCORE :::: " , lr_scaled_penalty_elasticnet.score(X_train_scaled, y_train))
print("X_test  SCORE :::: " , lr_scaled_penalty_elasticnet.score(X_test_scaled, y_test))
```

#### 스케일링 X

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)

print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

```python
# L1 규제 모델
lr_penalty_l1 = LogisticRegression(penalty='l1', solver='liblinear')
lr_penalty_l1.fit(X_train, y_train)

print("X_train SCORE :::: " , lr_penalty_l1.score(X_train, y_train))
print("X_test  SCORE :::: " , lr_penalty_l1.score(X_test, y_test))
```

```python
# L2 규제 모델
lr_penalty_l2 = LogisticRegression(penalty='l2')
lr_penalty_l2.fit(X_train, y_train)

print("X_train SCORE :::: " , lr_penalty_l2.score(X_train, y_train))
print("X_test  SCORE :::: " , lr_penalty_l2.score(X_test, y_test))
```

```python
# Elastic Net 규제 모델
lr_penalty_elasticnet = LogisticRegression(penalty='elasticnet', solver='saga', l1_ratio=0.5)
lr_penalty_elasticnet.fit(X_train, y_train)

print("X_train SCORE :::: " , lr_penalty_elasticnet.score(X_train, y_train))
print("X_test  SCORE :::: " , lr_penalty_elasticnet.score(X_test, y_test))
```

```python
# L1 규제 모델
lr_penalty_l1 = LogisticRegression(penalty='l1', solver='liblinear')
lr_penalty_l1.fit(X_train, y_train)

print("X_train SCORE :::: " , lr_penalty_l1.score(X_train, y_train))
print("X_test  SCORE :::: " , lr_penalty_l1.score(X_test, y_test))
```

```python
model = LogisticRegression(penalty='l2', solver='newton-cg')
model.fit(X_train, y_train)
print("X_train SCORE :::: " , model.score(X_train, y_train))
print("X_test  SCORE :::: " , model.score(X_test, y_test))
```

## 결정 트리 모델

#### 스케일링 O

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_scaled, y_train)

print(dt.score(X_train_scaled, y_train))
print(dt.score(X_test_scaled, y_test))
```

#### 스케일링 X

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

print(dt.score(X_train, y_train))
print(dt.score(X_test, y_test))
```

```python
# plot_tree() 시각화
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(dt)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(10,7))
plot_tree(dt, max_depth=10, filled=True, feature_names=['Pclass','Age','Fare','SibSp','Parch','Sex_male'])
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

for i in range(1,10):
    dt = DecisionTreeClassifier(max_depth=i, random_state=42)
    dt.fit(X_train_scaled, y_train)

    print("Depth :::: ", i, " // train_score :::: ", dt.score(X_train_scaled, y_train))
    print("Depth :::: ", i, " // test_score  :::: ", dt.score(X_test_scaled, y_test))
    print("Score gap :::: " ,  dt.score(X_train_scaled, y_train)- dt.score(X_test_scaled, y_test))    
    print()    
```

```python
# 그래프로 확인
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 15))
plot_tree(dt, filled=True, feature_names=['Pclass','Age','Fare','SibSp','Parch','Sex_male'])
plt.show()
```

```python
# 특성 중요도 ['Pclass','Age','Fare','SibSp','Parch','Sex_male']
dt.feature_importances_
```

```python
# min_impurity_decrease > 하이퍼 파라미터
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(min_impurity_decrease=0.005, random_state=42)
dt.fit(X_train, y_train)

print(dt.score(X_train, y_train))
print(dt.score(X_test, y_test))

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(20,15))
plot_tree(dt, filled=True, feature_names=['Pclass','Age','Fare','SibSp','Parch','Sex_male'])
plt.show()
```

```python
# min_impurity_decrease > 하이퍼 파라미터
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42, max_depth=5, min_samples_split=10, max_features=2)
dt.fit(X_train, y_train)

print(dt.score(X_train, y_train))
print(dt.score(X_test, y_test))
```

## 평가 지표 확인

```python
y_pred = dt.predict(X_test)
print(y_pred)
```

```python
y_test
```

```python
y_pred == y_test
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print(f'ROC-AUC: {roc_auc}')
print('Confusion Matrix:')
print(conf_matrix)
```
